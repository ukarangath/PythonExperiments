```python
  question_template = {
    "title": "default title",
    "question": "default question",
    "answer": "default answer",
    "hints": []
}
def make_new_question(title, question, answer, hints=None):
    new_q = question_template.copy()

    # always require title, question, answer
    new_q["title"] = title
    new_q["question"] = question
    new_q["answer"] = answer

    # sometimes there aren't hints, that's fine. Otherwise, add them:
    if hints is not None:
        new_q["hints"].extend(hints)

    return new_q
```

```python
# Password check of python
def interpretCashier(cashier_entry_str):
    if(len(cashier_entry_str) == 0): 
        return ("Got None") 
    else : 
        if cashier_entry_str.isdigit():
            return "PIN"
        elif '.' in cashier_entry_str:
            temp = cashier_entry_str.replace('.','')
            if temp.isnumeric():
                return 'Transaction'
            else:
                return 'Password'
        else:
            return 'Password'
```

```python
print(interpretCashier('3323242344'))
print(interpretCashier('dqdqwdqwdqwdddwdwqw323131;;;'))
print(interpretCashier('232.32323'))
print(interpretCashier('++++++######'))
```

```python
def find_color(strRGB):
    if strRGB is None:
        return None
    # Removing the first 'rgb(' part and last ')' part of the string
    # Starts from 4th literal(from 0th literal) till -1th literal(last literal of string), but not inclduing it
    strRGB = strRGB[4:-1] 
    # Replaces comma with empty character, means removing commas from string
    strRGB = strRGB.replace(',', '')
    # Getting remaining numbers after splitting on space
    listRGB = [int(color) for color in strRGB.split(" ")]
    red = listRGB[0]
    green = listRGB[1]
    blue = listRGB[2]
    if red > green and red > blue:
        return 'red'
    if green > red and green > blue:
        return 'green'    
    if blue > green and blue > red:
        return 'blue'
    if red == green and green == blue:
        return 'grey'
    if red == green:
        return 'yellow'
    if red == blue:
        return 'purple'
    if green == blue:
        return 'teal'
```

```python
def string_splitter(input_string):
    if len(input_string) == 0:
        return []    
    if input_string[0] == ' ':
        input_string = input_string[1:]
    if input_string[len(input_string)-1] == ' ':
        input_string = input_string[0:len(input_string)-1]
    if len(input_string) == 0:
        return []
    print('*',input_string,'*',sep='')
    result_list = []
    split_locations = []
    for index in range(len(input_string)):
        if input_string[index] == ' ':
            split_locations = split_locations + [index]
    print(split_locations)
    result_list = result_list + [input_string[0:split_locations[0]]]
    for index in range(len(split_locations)):
        if index > 0:
            begin = split_locations[index-1] + 1
            end = split_locations[index]
            result_list = result_list + [input_string[begin:end]]
    begin = split_locations[len(split_locations)-1] + 1
    end = len(input_string)
    result_list = result_list + [input_string[begin:end]]
    return result_list
```

```python
st = 'This article also discussed that problem'
print(len(st))
print(string_splitter(st))
```

```python
def insertionSort(array):
    # Write your code here.
    for Idx in range(len(array)):
        insertAndSettle(array, Idx)
    return array
def insertAndSettle(array, Idx):
    for Idx in range(Idx, -1, -1):
        if Idx - 1 < 0:
            break
        if array[Idx] >= array[Idx-1]:
            break
        else:
            array[Idx], array[Idx-1] = array[Idx-1], array[Idx]
```

```python
array = [9, 5, 1, 4, 4, 3]
insertionSort(array)
```
